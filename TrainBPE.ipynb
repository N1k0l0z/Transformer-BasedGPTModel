```python
with open('C:\\Users\\NiKordzakhia\\Desktop\\Transformer-BasedGPTModel\\vefxistyaosani.txt', 'r') as f:
    data = f.read()
```

```python
tokenizer = BPETOKENIZER(data, 7000)
tokenizer.train()
```

```python
with open('C:\\Users\\NiKordzakhia\\Desktop\\Transformer-BasedGPTModel\\tokenizer', 'wb') as f:
    pickle.dump(tokenizer, f)
with open('C:\\Users\\NiKordzakhia\Desktop\\Transformer-BasedGPTModel\\tokenizer', 'rb') as f:
    tokenizer = pickle.load(f)
with open('C:\\Users\\NiKordzakhia\\Desktop\\Transformer-BasedGPTModel\\vefxistyaosani.txt', 'r') as f:
    data = f.read()
```

```python
encoded = tokenizer.encode(data[:1000])
for i in encoded:
    print(tokenizer.decode([i]), end='\n')
```
